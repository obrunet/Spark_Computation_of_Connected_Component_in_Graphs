```python
import pandas as pd
import numpy as np

import seaborn as sns
```

```python
df_notre_dame = pd.DataFrame({
    'Cluster': ['Databricks', 'G.C.P', 'LAMSADE'],
    'RDD': [333, 272, 122],
    'DF': [379, 168, 114]
})

df_google = pd.DataFrame({
    'Cluster': ['Databricks', 'G.C.P', 'LAMSADE'],
    'RDD': [1012, 497, 132],
    'DF': [1165, 425, 222]
})
```

```python
df_notre_dame = pd.melt(df_notre_dame, id_vars=['Cluster'], value_vars=['RDD', 'DF'])
df_notre_dame.rename(columns={'variable': 'RDD/DF', 'value': 'computation time'}, inplace=True)
df_notre_dame
```

```python
df_google = pd.melt(df_google, id_vars=['Cluster'], value_vars=['RDD', 'DF'])
df_google.rename(columns={'variable': 'RDD/DF', 'value': 'computation time'}, inplace=True)
df_google
```

```python
sns.barplot(data=df_notre_dame, x="Cluster", y="computation time", hue="RDD/DF").set(title='Data set web-NotreDame - Computation time (the lower the better)')
```

```python
sns.barplot(data=df_google, x="Cluster", y="computation time", hue="RDD/DF").set(title='Data set web-Google - Computation time (the lower the better)')
```
